# **Outlier Detection**

```python
df_melted = pd.melt(df, id_vars=["month_number", "total_profit"],
                     value_vars=['facecream', 'facewash', 'toothpaste',
                                 'bathingsoap', 'shampoo', 'moisturizer'],
                     var_name='Product', value_name='Sales')

# Calculate Q1, Q3, and IQR
Q1_sales = df_melted.groupby('Product')['Sales'].quantile(0.25)
Q3_sales = df_melted.groupby('Product')['Sales'].quantile(0.75)
IQR_sales = Q3_sales - Q1_sales

# Define outliers
outliers = df_melted[(df_melted['Sales'] < (df_melted['Product'].map(Q1_sales) - 1.5 * df_melted['Product'].map(IQR_sales))) |
                      (df_melted['Sales'] > (df_melted['Product'].map(Q3_sales) + 1.5 * df_melted['Product'].map(IQR_sales)))]

# Plotting boxplots for each product
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product', y='Sales', data=df_melted)
plt.title('Boxplot of Sales for Each Product')
plt.show()

# Display the number of outliers detected for each product
outliers_count = outliers['Product'].value_counts()
print("Number of outliers detected for each product:")
print(outliers_count)
```

```python
df_cleaned = df_melted[~((df_melted['Sales'] < (df_melted['Product'].map(Q1_sales) - 1.5 * df_melted['Product'].map(IQR_sales))) |
                         (df_melted['Sales'] > (df_melted['Product'].map(Q3_sales) + 1.5 * df_melted['Product'].map(IQR_sales))))]

# Plot the boxplots for each product after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product', y='Sales', data=df_cleaned)
plt.title('Boxplot of Sales for Each Product (Outliers Removed)')
plt.show()

# Display the number of data points remaining for each product after removing outliers
print("Number of data points after removing outliers:")
print(df_cleaned['Product'].value_counts())
```

# **Line Plot (Month vs Total Profit)**

```python
plt.figure(figsize=(10, 5))
plt.plot(df["month_number"], df["total_profit"], marker = 'o', color = 'b')
plt.title("Total Profit of all Months")
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
```

# **Multiline Plot (Product Sales Data)**


```python
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['facecream'], marker = 'o', label = 'Face Cream Sales')
plt.plot(df['month_number'], df['facewash'], marker = 'o', label = 'Face Wash Sales')
plt.plot(df['month_number'], df['toothpaste'], marker = 'o', label = 'Toothpaste Sales')
plt.plot(df['month_number'], df['bathingsoap'], marker = 'o', label = 'Bathing Soap Sales')
plt.plot(df['month_number'], df['shampoo'], marker = 'o', label = 'Shampoo Sales')
plt.plot(df['month_number'], df['moisturizer'], marker = 'o', label = 'Moisturizer Sales')
plt.title('Product Sales Data')
plt.xlabel("Month")
plt.ylabel('Sales Unit')
plt.legend()
plt.show()
```

# **Scatter Plot of Toothpaste Sales Data**

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['month_number'], df['toothpaste'], color = 'r')
plt.xlabel("Month")
plt.ylabel("Toothpaste Sales")
plt.show()
```

# **Bar Chart of Face Cream and Face Wash Product Sales**

```python
df.plot(x='month_number',y=['facecream','facewash'],kind='bar',color=['blue','red'])
plt.show()
```

```python
sns.histplot(df['total_profit'],color='red')
plt.show()
```
