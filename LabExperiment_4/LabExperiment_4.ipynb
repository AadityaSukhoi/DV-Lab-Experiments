```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AirPassengers.csv')
```

```python
df.head()
```

```python
# sort the data based on the month

df['Month'] = pd.to_datetime(df['Month'])
```

```python
# find the missing values in the data

df.isnull().sum()
```

```python
#  Plot the data to see the trend in the data

plt.plot(df['Month'], df['#Passengers'])
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers over time')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['#Passengers'], vert=False, patch_artist=True)
plt.xlabel('Number of Passengers')
plt.title('Boxplot of Number of Passengers')
plt.show()
```

```python
# Convert 'Month' to datetime if not already done
df['Month'] = pd.to_datetime(df['Month'])

# Extract month name or number
df['MonthOnly'] = df['Month'].dt.month  # Use .dt.month_name() for month names

# Create a boxplot for the number of passengers by month
plt.figure(figsize=(12, 6))
sns.boxplot(x='MonthOnly', y='#Passengers', data=df)
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.title('Monthly Distribution of Number of Passengers')

plt.show()
```

```python
# Ensure 'Month' is in datetime format
df['Month'] = pd.to_datetime(df['Month'])

# Extract the year
df['Year'] = df['Month'].dt.year

# Create a boxplot for the number of passengers by year
plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='#Passengers', data=df)
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Yearly Distribution of Number of Passengers')

plt.show()
```

```python
# Create a pie chart
# Ensure 'Month' is in datetime format
df['Month'] = pd.to_datetime(df['Month'])

# Extract the year
df['Year'] = df['Month'].dt.year

# Aggregate the number of passengers by year
passengers_by_year = df.groupby('Year')['#Passengers'].sum()

# Create a pie chart for the total passengers by year
plt.figure(figsize=(8, 8))
plt.pie(passengers_by_year, labels=passengers_by_year.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Passenger Distribution by Year')

plt.show()
```

```python
# Extract the month
df['MonthOnly'] = df['Month'].dt.month

# Aggregate the number of passengers by month
passengers_by_month = df.groupby('MonthOnly')['#Passengers'].sum()

# Create a pie chart for the total passengers by month
plt.figure(figsize=(8, 8))
plt.pie(passengers_by_month, labels=passengers_by_month.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Passenger Distribution by Month')

plt.show()
```
