# **Outlier Detection and Removal**

```python
#using z-score
from scipy import stats
numerical_columns = ['result_margin', 'target_runs', 'target_overs']

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df[numerical_columns]))

# Define a threshold for Z-scores (e.g., 3 standard deviations)
threshold = 3
outliers = np.where(z_scores > threshold)

# Print counts of outliers for each column
for col in numerical_columns:
    num_outliers = len(df[(np.abs(stats.zscore(df[col])) > threshold)])
    print(f"Number of outliers in {col}: {num_outliers}")

# Plotting the outliers using boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_columns])
plt.title("Boxplot for Outlier Detection")
plt.show()
```

# **Outlier Removal**

```python
#using z-score
from scipy import stats
numerical_columns = ['result_margin', 'target_runs', 'target_overs']

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df[numerical_columns]))

# Define a threshold for Z-scores (e.g., 3 standard deviations)
threshold = 3
outliers = np.where(z_scores > threshold)

# Print counts of outliers for each column
for col in numerical_columns:
    num_outliers = len(df[(np.abs(stats.zscore(df[col])) > threshold)])
    print(f"Number of outliers in {col}: {num_outliers}")

# Plotting the outliers using boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_columns])
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
#drop outliers using the z-score
mask = (z_scores < threshold).all(axis=1)

# Drop outliers from the DataFrame
df_cleaned = df[mask]

print(f"DataFrame shape after removing outliers using Z-score method: {df_cleaned.shape}")
```

# **Plot to visualize the number of matches held in each city**

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='city', order=df['city'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Matches Held in Each City')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```

# **Plot to visualize the number of matches won by each team.**

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='winner', order=df['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Matches Won')
plt.show()
```

# **Plot to visualize the number of matches held every season.**

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='season', order=df['season'].value_counts().index)
plt.title('Number of Matches Held Every Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.show()
```

# **Plot to visualize the top 10 players based on the number of MOM (Man of the Match) awards won.**

```python
top_players = df['player_of_match'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_players.values, y=top_players.index)
plt.title('Top 10 Players Based on the Number of MOM Awards Won')
plt.xlabel('Number of MOM Awards')
plt.ylabel('Player')
plt.show()
```
